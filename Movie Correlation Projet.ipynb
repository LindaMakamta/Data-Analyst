```python
#import librairie

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize']= (12,8) #ajuster la configuration de plot



#lire les donnée 

df = pd.read_csv(r'C:\Users\sorel\OneDrive\Bureau\Sql\movies\movies.csv')


```

```python
# regardon les donnée 

df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%' .format(col, pct_missing))
```

```python
#type de donnée des colonne 

df.dtypes
```

```python
#changement de type

df['budget'] = df['budget'].fillna(0).astype('int64')

df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df.head()
```

```python
#creer une colonne d'année correct
df['yearcorrect'] = df['released'].astype(str).str[:4]

df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
#supprimer tous les doublons

df['company'].drop_duplicates().sort_values(ascending= False)
```

```python
df= df.sort_values(by= ['gross'],inplace = False, ascending= False)
```

```python
 df= df.sort_values(by= ['gross'],inplace = False, ascending= False)
#budget de forte correlation
#companie de forte correlation
```

```python
#nuage de point du budget par rapport au revenu brut


plt.scatter(x=df['budget'], y= df['gross'])
plt.title('Budget Vs Gross')
plt.xlabel('Gross')
plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
#tracer le trait de regression avec seaborn

sns.regplot(x='budget', y= 'gross', data= df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#pearson, kendall,spearman
df.corr(method= 'spearman')
```

```python
correlation_matrix = df.corr(method= 'pearson')
sns.heatmap(correlation_matrix, annot= True)

plt.title('Matrice De Correlation Pour Les Caracteristique Numerique')
plt.xlabel('Caracteristique Du Filme')
plt.ylabel('Caracteristique Du Filme')
plt.show()
```

```python
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method= 'pearson')
sns.heatmap(correlation_matrix, annot= True)

plt.title('Matrice De Correlation Pour Les Caracteristique Numerique')
plt.xlabel('Caracteristique Du Filme')
plt.ylabel('Caracteristique Du Filme')
plt.show()
```

```python
df_numerized.corr().head()
```

```python
correlation_mat = df_numerized.corr()

corr_pair = correlation_mat.unstack()

corr_pair
```

```python
sorted_pairs = corr_pair.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
